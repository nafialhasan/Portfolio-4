```python
# importing necessary libraries

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')
```

#### Selecting features for further analysis
Selecting the five most correlated features ('Doors', 'Mileage', 'Category', 'Gear box type', and 'Wheel') for further analysis is reasonable as it focuses on the most relevant and informative features for predicting the target variable ('Price'). This can lead to improved model performance, reduced overfitting, and better interpretability.

```python
# creating a box plot for five most correlated features

plt.figure(figsize = (10, 6))
car_data[["Doors", "Mileage", "Category", "Gear box type", "Wheel"]].boxplot()
plt.title("Box Plot of Five Most Correlated Features")
plt.show()
```

#### Observation:

'Doors', 'Gear box type', and 'Wheel' have outliers in their respective records. Removing outliers from these features before further analysis and model building is needed for the following reasons:
- Improved data quality by eliminating errors or anomalies.
- Reduced bias and skewness in feature distributions.
- Enhanced model performance and generalization by mitigating the disproportionate influence of outliers.
- Increased interpretability of feature-target relationships.
- Robustness to extreme values that could distort predictions.

```python
# calculating the interquartile range (iqr)

doors_q1 = car_data["Doors"].quantile(0.25)
doors_q3 = car_data["Doors"].quantile(0.75)
doors_iqr = doors_q3 - doors_q1

gear_box_type_q1 = car_data["Gear box type"].quantile(0.25)
gear_box_type_q3 = car_data["Gear box type"].quantile(0.75)
gear_box_type_iqr = gear_box_type_q3 - gear_box_type_q1

wheel_q1 = car_data["Wheel"].quantile(0.25)
wheel_q3 = car_data["Wheel"].quantile(0.75)
wheel_iqr = wheel_q3 - wheel_q1

# setting the lower and upper bound

doors_lower_bound = doors_q1 - 1.5 * doors_iqr
doors_upper_bound = doors_q3 + 1.5 * doors_iqr

gear_box_type_lower_bound = gear_box_type_q1 - 1.5 * gear_box_type_iqr
gear_box_type_upper_bound = gear_box_type_q3 + 1.5 * gear_box_type_iqr

wheel_lower_bound = wheel_q1 - 1.5 * wheel_iqr
wheel_upper_bound = wheel_q3 + 1.5 * wheel_iqr

# finding the outliers

doors_outliers = car_data[(car_data["Doors"] < doors_lower_bound) | (car_data["Doors"] > doors_upper_bound)]

gear_box_type_outliers = car_data[(car_data["Gear box type"] < gear_box_type_lower_bound) | (car_data["Gear box type"] > gear_box_type_upper_bound)]

wheel_outliers = car_data[(car_data["Wheel"] < wheel_lower_bound) | (car_data["Wheel"] > wheel_upper_bound)]

# combining the outliers

outliers = pd.concat([doors_outliers, gear_box_type_outliers, wheel_outliers])
outliers = outliers.drop_duplicates()

# printing the number of outliers

print("The number of outliers in the dataset: ", outliers.shape[0])
```

```python
# removing the outliers from the dataset

cleaned_car_data = car_data.drop(outliers.index)
```

```python
# number of records after removing outliers

print("The number of records after removing the outliers:", len(cleaned_car_data))
```

## Splitting the dataset

```python
# spliting the data into train and test set

train, test = train_test_split(cleaned_car_data, test_size = 0.25, random_state = 42)

# printing the number of records in the splits

print("Number of records in trining set: ", train.shape[0])
print("Number of records in testing set: ", test.shape[0])
```

#### Observation:
The train set has 12,383 records and test set has 4,128 records, after splitting the dataset into 0.75 : 0.25 ratio

```python
# building model with the five most correlated features

x_train = train[["Doors", "Mileage", "Category", "Gear box type", "Wheel"]]
y_train = train["Price"]
x_test = test[["Doors", "Mileage", "Category", "Gear box type", "Wheel"]]
y_test = test["Price"]
```

## Building Linear Regression Model

#### Reasons behind choosing Linear Regression to build a model

The target variable 'Price' is a continuous numerical variable, which makes it a regression problem. Linear regression is a widely used and well-established technique for modeling continuous target variables.

```python
# creating a linear regression model

lin_reg = LinearRegression()

# training the model with the selected features

lin_reg.fit(x_train, y_train)
```

```python
# predicting based on the test set with the linear regression model

y_lin_reg_predicted = lin_reg.predict(x_test)

# printing the predictions

print("Predictions: ", y_lin_reg_predicted)
```

```python
# finding the residuals

lin_reg_residuals = y_test - y_lin_reg_predicted

# plotting scatter graph to compare the residuals and predicated values

plt.scatter(y_lin_reg_predicted, lin_reg_residuals)
plt.axhline(y = 0, color = "r", linestyle = "--")
plt.xlabel("Linear Reg Predicted Values")
plt.ylabel("Linear Reg Residuals")
plt.title("Linear Reg Residuals vs. Linear Reg Predicted Values")
plt.show()
```

#### Observation:
The scatter plot titled “Linear Reg Residuals vs. Linear Reg Predicted Values” illustrates the residuals of a linear regression model against the predicted values. The majority of data points are clustered around the zero line on the y-axis, indicating accurate predictions with small residuals. However, some points with large residuals suggest inaccuracies in the model’s predictions for certain data points.

```python
# evaluating the model

lin_reg_mae = mean_absolute_error(y_test, y_lin_reg_predicted)
lin_reg_mse = ((np.array(y_test) - y_lin_reg_predicted) ** 2).sum() / len(y_test)
lin_reg_rmse = np.sqrt(lin_reg_mse)
lin_reg_r2 = r2_score(y_test, y_lin_reg_predicted)

print("Mean Absolute Error of the Linear Regression Model: ", lin_reg_mae)
print("Mean Squared Error of the Linear Regression Model: ", lin_reg_mse)
print("Rooted Mean Squared Error of the Linear Regression Model: ", lin_reg_rmse)
print("R-Squared of the Linear Regression Model: ", lin_reg_r2)
```

#### Analysis
- The MAE value of 12929.92 indicates that, on average, the linear regression model's predictions deviate from the actual target values by approximately $12,929.
- The MSE value of 574357415.89 is quite high, indicating that the model's predictions deviate significantly from the actual target values.
- The RMSE value of 23965.76 is also high, suggesting that the model's predictions have a substantial average error.
- The R2 value of 0.0433 (or 4.33%) is extremely low, indicating that the model explains only a small portion of the variance in the 'Price' variable.

## Building Polynomial Regression Model

```python
# creating a pipeline with polynomial features and linear regression

poly_reg = Pipeline([('poly', PolynomialFeatures(degree = 2)), ('lin_reg', LinearRegression())])

# training the polynomial regression model

poly_reg.fit(x_train, y_train)
```

```python
# predicting based on the test set with the polynomial regression model

y_poly_reg_predicted = poly_reg.predict(x_test)

# printing the predictions

print("Predictions: ", y_poly_reg_predicted)
```

```python
# finding the residuals

poly_reg_residuals = y_test - y_poly_reg_predicted

# plotting scatter graph to compare the residuals and predicated values

plt.scatter(y_poly_reg_predicted, poly_reg_residuals)
plt.axhline(y = 0, color = "r", linestyle = "--")
plt.xlabel("Polynomial Reg Predicted Values")
plt.ylabel("Polynomial Reg Residuals")
plt.title("Polynomial Reg Residuals vs. Polynomial Reg Predicted Values")
plt.show()
```

#### Observation:
The scatter plot titled “Polynomial Reg Residuals vs. Polynomial Reg Predicted Values” shows the residuals of a polynomial regression model against the predicted values. The majority of data points cluster near the zero line, indicating small residuals for many predictions. However, there are outliers with large residuals, suggesting inaccuracies in the model’s predictions for certain data points.

```python
# evaluating the model

poly_reg_mae = mean_absolute_error(y_test, y_poly_reg_predicted)
poly_reg_mse = ((np.array(y_test) - y_poly_reg_predicted) ** 2).sum() / len(y_test)
poly_reg_rmse = np.sqrt(poly_reg_mse)
poly_reg_r2 = r2_score(y_test, y_poly_reg_predicted)

print("Mean Absolute Error of the Polynomial Regression Model: ", poly_reg_mae)
print("Mean Squared Error of the Polynomial Regression Model: ", poly_reg_mse)
print("Rooted Mean Squared Error of the Polynomial Regression Model: ", poly_reg_rmse)
print("R-Squared of the Polynomial Regression Model: ", poly_reg_r2)
```

#### Analysis
- The MAE value of 12475.30 indicates that, on average, the polynomial regression model's predictions deviate from the actual target values by approximately $12,475.
- The MSE value of 561383388.00 is slightly lower than the MSE of the linear regression model (574357415.89), indicating a marginal improvement in the model's predictions.
- The RMSE value of 23693.53 is also slightly lower than the RMSE of the linear regression model (23965.76). This indicates a small reduction in the average error of the model's predictions.
- The R2 value of 0.0649 (or 6.49%) is higher than the R2 of the linear regression model (0.0433 or 4.33%), indicating that the polynomial regression model explains a slightly larger portion of the variance in the 'Price' variable.

The polynomial regression model shows a slight improvement over the linear regression model, with a marginally lower MSE and RMSE, a slightly higher R2 value, and a lower MAE. However, the model's overall performance remains poor, with high error values (MSE, RMSE, and MAE) and a low ability to explain the variance in the 'Price' variable (low R2). While the polynomial model is better, both models struggle to accurately predict car prices based on the given features.
The lower MAE value for the polynomial regression model suggests that its predictions have a slightly smaller average magnitude of error compared to the linear regression model. However, the improvement is marginal, and the model still exhibits high error values and a low ^2 score, indicating poor overall performance.


## Building Decision Tree Regression Model

```python
# creating the decision tree regression model

dt_reg = DecisionTreeRegressor(random_state = 42)

# training the model

dt_reg.fit(x_train, y_train)
```

```python
# predicting based on the test set with the decision tree regression model

y_dt_reg_predicted = dt_reg.predict(x_test)

# printing the predictions

print("Predictions: ", y_dt_reg_predicted)
```

```python
# finding the residuals

dt_reg_residuals = y_test - y_dt_reg_predicted

# plotting scatter graph to compare the residuals and predicated values

plt.scatter(y_dt_reg_predicted, dt_reg_residuals)
plt.axhline(y = 0, color = "r", linestyle = "--")
plt.xlabel("Decision Trees Reg Predicted Values")
plt.ylabel("Decision Trees Reg Residuals")
plt.title("Decision Trees Reg Residuals vs. Decision Trees Reg Predicted Values")
plt.show()
```

#### Observation:
The scatter plot titled “Decision Trees Reg Residuals vs. Decision Trees Reg Predicted Values” shows the residuals of a decision tree regression model against the predicted values. Most data points are clustered around the zero line, indicating small residuals and accurate predictions for many instances. However, there are outliers with large residuals, suggesting some predictions are significantly off. This indicates the model’s limitations in capturing all the nuances of the dataset.

```python
# evaluating the model

dt_reg_mae = mean_absolute_error(y_test, y_dt_reg_predicted)
dt_reg_mse = ((np.array(y_test) - y_dt_reg_predicted) ** 2).sum() / len(y_test)
dt_reg_rmse = np.sqrt(dt_reg_mse)
dt_reg_r2 = r2_score(y_test, y_dt_reg_predicted)

print("Mean Absolute Error of the Decision Trees Regression Model: ", dt_reg_mae)
print("Mean Squared Error of the Decision Trees Regression Model: ", dt_reg_mse)
print("Rooted Mean Squared Error of the Decision Trees Regression Model: ", dt_reg_rmse)
print("R-Squared of the Decision Trees Regression Model: ", dt_reg_r2)
```

#### Analysis
- The MAE value of 11005.63 indicates that, on average, the decision tree regression model's predictions deviate from the actual target values by approximately $11,005.
- The MSE value of 620547910.19 is higher than both the linear regression (574357415.89) and polynomial regression (561383388.00) models. This indicates that the decision tree regression model's predictions deviate more from the actual target values compared to the other two models.
- The RMSE value of 24910.80 is also higher than the RMSE values of the linear regression (23965.76) and polynomial regression (23693.53) models. This suggests that the decision tree regression model has a higher average error in its predictions.
- The R2 value of -0.0337 (or -3.37%) is negative, which is an unusual and undesirable outcome. A negative R2 value indicates that the model performs worse than simply using the mean of the target variable as a constant prediction.

Overall, the decision tree regression model exhibits poorer performance compared to both the linear regression and polynomial regression models for this dataset. The higher MSE and RMSE values, coupled with the negative R2 value and the higher MAE, suggest that the decision tree regression model is not able to capture the underlying patterns and relationships between the features and the target variable ('Price') effectively.
The decision tree regression model has the lowest MAE among the three models, indicating that its predictions have a smaller average magnitude of error compared to the actual target values. However, the negative R2 value and higher MSE and RMSE values compared to the other models suggest that the decision tree regression model may not be the most suitable choice for this dataset.
